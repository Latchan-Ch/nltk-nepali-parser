```python
import nltk

# Define the Context-Free Grammar for Nepali (Subject-Object-Verb)
# We are creating a small dictionary (lexicon) of Nepali words to test.
nepali_grammar = nltk.CFG.fromstring("""
  S -> NP VP
  VP -> NP V
  NP -> Noun

  # Lexicon (Vocabulary)
  Noun -> 'रामले' | 'सिताले' | 'स्याउ' | 'किताब' | 'भात'
  V -> 'खायो' | 'पढ्यो' | 'खाइन्'
""")

# Initialize the parser with our grammar
parser = nltk.ChartParser(nepali_grammar)
print(" Nepali Parser is ready!")
```

```python
# Let's test a simple Nepali sentence: "Ram apple ate" (रामले स्याउ खायो)
test_sentence = "रामले स्याउ खायो"

# Split the sentence into a list of words
tokens = test_sentence.split()

print(f"Parsing sentence: {test_sentence}\n")

# Parse the sentence and print the tree
try:
    for tree in parser.parse(tokens):
        # pretty_print() works perfectly in Google Colab!
        tree.pretty_print()
except ValueError as e:
    print(f"Error: A word in the sentence is not in our dictionary. {e}")
```

```python
# A list of test sentences (some correct, some grammatically wrong, some with unknown words)
test_dataset = [
    "रामले स्याउ खायो",       # Valid
    "सिताले किताब पढ्यो",       # Valid
    "रामले भात खायो",         # Valid
    "स्याउ रामले खायो",       # Invalid structure (Object-Subject-Verb)
    "हरिले स्याउ खायो"        # 'हरिले' is not in our vocabulary
]

successful_parses = 0

print("--- Model Evaluation ---")
for sentence in test_dataset:
    tokens = sentence.split()
    parsed = False
    try:
        # Check if the parser can find at least one valid tree
        trees = list(parser.parse(tokens))
        if len(trees) > 0:
            parsed = True
            successful_parses += 1
    except ValueError:
        pass # Word not in vocabulary

    print(f"Sentence: '{sentence}' -> Parsed Successfully: {parsed}")

# Calculate simple accuracy based on our dataset
accuracy = (successful_parses / len(test_dataset)) * 100
print(f"\nModel Accuracy on Test Set: {accuracy}%")
```
